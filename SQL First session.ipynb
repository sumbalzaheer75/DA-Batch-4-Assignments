### Ascending/Descending order by

```python
q = '''select jobtitle Title, TotalPay Pay from salaries
    order by 2'''

sql_query(q)
```

```python
q = '''select jobtitle Title, TotalPay Pay from salaries
    order by totalpay'''

sql_query(q)
```

```python
q = '''select jobtitle Title, TotalPay Pay from salaries
    order by 4'''

sql_query(q)
```

```python
q = '''select jobtitle Title, TotalPay Pay from salaries
    order by 2 desc'''

sql_query(q)
```

```python
q = '''select jobtitle Title, TotalPay Pay from salaries
    where jobtitle != 'Not provided'
    order by 2 desc'''

sql_query(q)
```

```python
q = '''select jobtitle Title, TotalPay Pay, OtherPay, BasePay as Base from salaries
    where jobtitle != 'Not provided'
    order by 4'''

sql_query(q)
```

```python
q = '''select jobtitle Title, TotalPay Pay, OtherPay, BasePay as Base from salaries
    where jobtitle != 'Not provided'
    order by basepay'''

sql_query(q)
```

```python
q = '''select jobtitle Title, TotalPay Pay, OtherPay, BasePay as Base from salaries
    where jobtitle != 'Not provided'
    order by base'''

sql_query(q)
```

```python
q = '''select jobtitle Title, TotalPay Pay from salaries order by 2 desc
    where jobtitle != 'Not provided'
    '''

sql_query(q)
```

### limit

```python
q = '''select employeename, totalpay from salaries
    order by 2 desc limit 10'''

sql_query(q)
```

```python
q = '''select employeename, totalpay limit 10 from salaries
    order by 2 desc'''

sql_query(q)
```

### operators

```python
# Show jobtitle title where overtimepay > 50k and excluded 'not provided' rows top 10 to descending
q = '''select jobtitle title, overtimepay from salaries
    where overtimepay > 50000
    and 
    overtimepay != 'Not Provided'
    order by 2 desc
    limit 10'''

sql_query(q)
```
